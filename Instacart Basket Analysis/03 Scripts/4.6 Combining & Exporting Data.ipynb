# 03. Merge both data frames

```python
# Perform a default test merge

pd.merge(df_ords_prods_comb,df_prods, on = ['product_id'], indicator = True)
```

```python
# Create new data frame with merged data

df_ords_prods_merged = df_ords_prods_comb.merge(df_prods, on = ['product_id'], indicator = True)
```
