# Analysis of results:

```python
# 4.8.6: Targeting customers by Spending pattern

# Aggregate using mean prices

ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Assign Spending Flag

ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High Spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# 4.8.6: There are 634,245 High Spenders; the rest are Low Spenders.

```python
# 4.8.7: Determing Frequent Customers

# Aggregate using median of ordering frequency

ords_prods_merge['ord_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Assign Spending Flag

ords_prods_merge.loc[ords_prods_merge['ord_freq'] <= 10, 'frequency_flag'] = 'Frequent Customer'
ords_prods_merge.loc[(ords_prods_merge['ord_freq'] > 10) & (ords_prods_merge['ord_freq'] <= 20), 'frequency_flag'] = 'Regular Customer'
ords_prods_merge.loc[ords_prods_merge['ord_freq'] > 20, 'frequency_flag'] = 'Non-Frequent Customer'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

# 4.8.7: There are 22790523 Frequent Customers.
