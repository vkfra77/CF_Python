```python
# Create Random sample, set Seed

np.random.seed(4)

# Create List with True/False values for the test np.random.rand() <= 0.7

dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# Store 70% of sample in data frame Big

Big = ords_prods_merge[dev]
```

```python
# Store 30% of sample in data frame Small

Small = ords_prods_merge[~dev]
```

```python
# Creating data frame with only necessary columns

df_2 = Small[['orders_day_of_week','prices']]
```

```python
# Creating Line Chart with Sample

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices',color = 'purple')
```

```python
# Export the visualization

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_30.png'))
```

```python
# Compare with Larger sample

df_4 = Big[['orders_day_of_week','prices']]
line_2 = sns.lineplot(data = df_4, x = 'orders_day_of_week',y = 'prices')
```

```python
# Export the visualization

line_2.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_70.png'))
```
